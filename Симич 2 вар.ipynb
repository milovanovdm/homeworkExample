4) Реализовать функцию perfect_number, которая принимает число и проверяет совершенное это число или нет. С помощью инструкции assert необходимо проверить тип данных. Совершенным числом является число, чья сумма его делителей, кроме него самого, равно ему. Пример 6 = 1+2+3

```python
def perfect_number(x):
    y = [j for j in range(1, x // 2 + 1) if not x % j]
    return sum(y) == x
    
a=input()
if a.isdigit() == True:
    a = int(a)
    print(perfect_number(a))
else:
    print ('ошибка')
```
